### 1. Variables in python
- A variable is a container for a value. It can be assigned a name; you can use it to refer to it later in the program. Based on the data type of the variable.
The interpreter allocates memory and decides what can be stored in the reserved memory. 

- Rules for variable names:
Not allowed naming conventions are:
    - Variable name cannot start with numbers (e.g. -> 0_a, 1_test )
    - White spaces (blanks) cannot be present in the variable name (e.g. -> a b, my variable). The most common practice is to fill the gap with an underscore character.
    - Special sign/symbol (“+” and “-“) cannot be used in variable name (e.g. -> a+b , x-y )
    - Python Reserve keyword cannot be used for naming a variable. (e.g. -> if, for, while etc)
    - Variables names are case-sensitive. (e.g.-> a and A is different)


```python
# Defind all the different types of variables
count = 100                       # Integer
distance = 1500.0                 # Float
name = "Hari"                     # String
z = 10j                           # Complex
print(count, distance, name, z)   # Print all the variable
```

### 4. Defining function

```python
# Write a code to generate fibonacci series till 10th term
def fibonacci(n):
    x1=0
    x2=1
    fibo = [x1, x2]
    for i in range(n-2):
        x3=x1+x2
        x1=x2
        x2=x3
        fibo.append(x3)
    return fibo
```

```python
x = fibonacci(10)
print(x)
```

```python
# create a function to generate factorial of n
def factorial(k):
    result = 1
    while k!=1:
        result = result*k
        k=k-1
    return result
```

```python
# Define a function that will take user input and calculate age. 
def calculate_age():
    # take user input  in ddmmyyyy format
    dob = input("Please type your age in ddmmyyyyy format:")
    # Extract year of birth using slicing technique
    yob = dob[4:8]
    # Type cast the year of birth to integer
    yob = int(yob)
    # Calculate age
    age = 2023-yob
    # print age
    print(age)

calculate_age()
```

```python
# Define a function that will take user input and advanec calculate age. 
def calculate_advance_age():
    # take user input  in ddmmyyyy format
    dob = input("Please type your age in ddmmyyyyy format:")

    # Extract year of birth using slicing technique
    yob = dob[4:8]
    mob = dob[2:4]
    day_b =dob[0:2]

    # Type cast the year of birth to integer
    yob = int(yob)
    mob = int(mob)
    day_b = int(day_b)

    # Convert to days
    birth_days = 365*yob +30*mob+day_b
    current_days = 365*2023 +30*10+28

    # Calculate age in days
    age_days = current_days - birth_days

    # convert to age in year
    age_year = age_days//365
    reminder_days = age_days%365
    age_month = reminder_days//30
    age_day = reminder_days%30
    
    print(f"The age is {age_year} year,{age_month} month, {age_day} days")

calculate_advance_age()
```
